# Police Incident Data (Safety)

## Description
We first selected the Pittsburgh Police Arrest Data as our initial metric to assess the safety of neighborhoods in Pittsburgh. We are assuming better neighborhoods have fewer arrests. To identify the 20 safest neighborhoods in terms of arrest rates, we utilized the value_counts() function to count the number of arrests for each unique neighborhood in the dataset. Subsequently, we applied the sort_values() function to rank the neighborhoods based on their relative safety levels.

```python
import pandas as pd
arrest_data = pd.read_csv("datasets/incidents.csv")
arrest_data.head()
```

```python
safety_ranking = arrest_data['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending = True)
top20_safety = safety_ranking.iloc[:20]
print(top20_safety)
```

## Graph
These two graphs show the 20 safest and most dangerous neighborhoods in Pittsburgh.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
top20_safety.plot(kind = 'bar', title = 'Top 20 safest neighborhood')
plt.xlabel("Neighborhood")
plt.ylabel("Number of Incident")
```

```python
incident_ranking = arrest_data ['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending = False)
top20_incident = incident_ranking.iloc[:20]
plt.figure(figsize=(10,10))
plt.pie(top20_incident, labels=top20_incident.index, autopct='%1.1f%%')
plt.title('20 Neighborhoods with the Most Incidents')
```

## Map
The following map shows the distribution of neighborhoods in Pittsburgh. Deeper color represents more arrests, meaning poor security in a neighborhood.

```python
import geopandas

neighborhoods = geopandas.read_file("neighborhoods/neighborhoods.shp")
arrest_map = neighborhoods.merge(
    safety_ranking,
    left_on='hood',
    right_index=True,
    how='left'
)
arrest_map[['hood','count','geometry']].head()

arrest_map.plot(column='count',
               cmap='OrRd',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Number of Incident"},
               figsize=(12, 8),
               missing_kwds={"color": "lightgrey"}
               )
```
